# skaters.csv

Contains skater (forward and defence) stats per season.

```python
from nhl_pool.config import PROCESSED_DIR
import pandas as pd

def stats_file_path(params):
    return PROCESSED_DIR / str(params["season"]) / params["season_type"] / f"{params["player_type"]}.csv"
```

```python
params = {
    "season": 20242025,
    "season_type": "regular",
    "player_type": "skaters"
}

path = stats_file_path(params)
skaters = pd.read_csv(path)
print(skaters.info())
skaters.head()
```

## How many players show up more than once?
Players could exist more than once due to trades.

```python
id_counts = skaters['playerId'].value_counts()

len(id_counts[id_counts > 1].index.tolist())
```

## What are the different positions?

```python
skaters["positionCode"].value_counts()
```

There aren't any cases where players play more than one position, probably just goes off their officially listed position.
