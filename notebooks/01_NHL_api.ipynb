## Importing the processed team codes

```python
from typing import Union, Any
from datetime import date

def build_season_query_key(year: Union[int, str]) -> str:
    year = int(year)
    return f"{str(year)}{str(year+1)}"
    
def package_params(team_dict, year, season_type):
    params = {
        "abbrev": team_dict.get("abbrev"),
        "season": build_season_query_key(year),
        "season_type": season_type
        }
    return params
```

```python
from nhl_pool.config import REFERENCE_DIR
from nhl_pool.dataset.fetch.fetch_team_codes import team_codes_raw_path
from nhl_pool.config import TEAM_CODES_REF_DIR

import json

with open(TEAM_CODES_REF_DIR, 'r') as f:
    team_dicts = json.load(f) 

for team_dict in team_dicts:
    params = package_params(team_dict, 2024, 2)
```

```python
print(params)
```

```python
def package_params_from_combination(combination):
    params = {
        "abbrev": combination[0],
        "season": build_season_query_key(combination[1]),
        "season_type": combination[2]
        }
    return params
```

```python
import itertools

abbrevs = [t["abbrev"] for t in team_dicts]
query_years = [2024]
query_season_types = [2, 3] # 1: preseason, 2: regular season, 3: playoffs


query_combinations = list(itertools.product(abbrevs, query_years, query_season_types))
query_combinations
```

```python
from nhl_pool.dataset.api import NHLAPI

params = package_params_from_combination(query_combinations[-1])
print(params)

api = NHLAPI()

data = api.get_team_roster_stats(params=params)
data
```

# Downloading game data

Need to get a list of all game ids. Can do this by getting the schedule of each team per season, which contains all of the game ids. It will then be compiled into a master list of game IDs. From this list, we can download the boxscore for each game.

```python
from nhl_pool.dataset.api import NHLAPI

api = NHLAPI()

params = {
    "abbrev": "ATL",
    "season": 20232024,
    "season_type": 2
}

data = api.get_team_schedule(params=params, force=False)
game_ids = [d["id"] for d in data['games']]
game_ids
```
