This is EDA on the data from the MoneyPuck source.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from pathlib import Path

BASE_PATH = Path(r"..\data\MoneyPuck\raw")
```

```python
def ffn_from_params(params):
    fpath = os.path.join(BASE_PATH, params["year"], params["season"])
    ffn = os.path.join(fpath, f"{params["category"]}.csv")
    return ffn
```

## Begin by investigating a single file (skater)

```python
params = {
    "year": "2024",
    "season": "regular",
    "category": "skaters"
    }

ffn = ffn_from_params(params)

df = pd.read_csv(ffn)
print(df.info())
df.head()
```

```python
list(df.columns)
```
